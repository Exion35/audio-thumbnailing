```python
# Bartsch, Mark A., and Gregory H. Wakefield. "Audio thumbnailing of popular music using chroma-based representations."
from ssm import ssm
import numpy as np
```

```python
def correlation_filtering(S):
    M, N = S.shape
    T = np.zeros(S.shape)
    k = 10
    for i in range(M - k):
        for j in range(N - i - k):
            for ki in range(k):
                T[i,j] += S[i + ki, (i + j) + ki]
    return T
```

```python
s = ssm('wav/sun.wav', treat = 0)
```

```python
import matplotlib.pyplot as plt
import librosa

T = correlation_filtering(s.s)
print(s.s.shape)
plt.figure(figsize=(12, 8))
librosa.display.specshow(T, x_axis='frames', y_axis='frames', n_xticks=12)
plt.title('SSM')
plt.set_cmap('hot_r')
plt.colorbar()
plt.show()
T.max()
```
